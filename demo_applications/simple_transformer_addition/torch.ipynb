```python
import torch
```

```python
x = torch.tensor([[[1,2,3], [4,5,6]]])
print(x)
print(x.shape)
print(x.ndim)
print(x.dtype)
print(x.device)
print(x.requires_grad)
print(x.grad)
print(x.grad_fn)
```

```python
print(x.squeeze())
print(x.squeeze(0))
print(x.squeeze(-1))
print(x.unsqueeze(0))
print(x.unsqueeze(1))
print(x.unsqueeze(-1))
```

```python
r = torch.randn(10, 10)
print(r)
print(r.shape)
print(r.ndim)
print(r.dtype)
print(r.device)
print(r.requires_grad)
print(r.grad)
print(r.unsqueeze(0))
```

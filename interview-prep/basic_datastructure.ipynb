## LRU CACHE

- Doubly linked list
- hashmap
- O(1) operations

```python
class Node:
    def __init__(self, k=0, v=0):
        self.key    =   k
        self.val    =   v
        self.prev   =   None
        self.next   =   None

class LRUCache:
    def __init__(self, capacity: int):
        self.cache    =   {}
        self.capacity =   capacity

        self.head   =   Node()
        self.tail   =   Node()
        self.head.next   =   self.tail
        self.tail.prev   =   self.head

    def remove(self, node):
        prev, next = node.prev, node.next
        prev.next   = next
        next.prev   =   prev

    def insert(self, node):
        prev, next = self.tail.prev, self.tail
        prev.next = next.prev = node
        node.prev, node.next = prev, next

    def get(self, key:  int) -> int:
        if key not in self.cache:
            return -1
        
        node = self.cache[key]
        self.remove(node)
        self.insert(node)

        return node.val

    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            self.remove(self.cache[key])
        
        node = Node(key, value)
        self.insert(node)
        self.cache[key] = node

        if len(self.cache) > self.capacity:
            lru = self.head.next
            self.remove(lru)


obj = LRUCache(2)
print(obj.put(1, 1))
print(obj.put(2, 2))
print(obj.get(1))
print(obj.put(3, 3))
print(obj.get(2))
print(obj.put(4, 4))
```

## Top K Frequent Elements

nums=[1,1,1,2,2,3]
k=2

output: [1,2]

```python
from collections import Counter
from typing import List, Any
import heapq

def topKFrequent_0(nums: List[int], k: int) -> List[int]:
    count = Counter[Any](nums)
    return heapq.nlargest(k, count.keys(), key=count.get)

def topKFrequent_1(nums: List[int], k: int) -> List[int]:
    count = Counter[Any](nums)
    heap = []
    for num, freq in count.items():
        heapq.heappush(heap, (freq, num))
        if len(heap) > k:
            heapq.heappop(heap)
    return [num for freq, num in heap]








```

## Merge Intervals
input:
[
 [1,3],
 [2,6],
 [8,10]
]
output:
[
 [1,6],
 [8,10]
]

```python
def merge_intervals(intervals: List[List[int]]) -> List[List[int]]:
    sorted_intervals = sorted(intervals, key=lambda x:x[0])
    merged = []
    for items in sorted_intervals:
        if not merged or merged[-1][1] < items[0]:
            merged.append(items)
        else:
            merged[-1][1] = max(items[1], merged[-1][1])
    
    return merged
    
```

## Transaction Fraud Window -  Sliding Window

Find users making more than:

3 transactions within 60 seconds

input:
[
("u1",10),
("u1",20),
("u1",50),
("u1",55)
]

```python
from collections import defaultdict, deque
from typing import List, Tuple

def transaction_fraud_window(transactions: List[Tuple[str, int]]) -> List[str]:
    users = defaultdict(deque)
    flagged = []
    for user, time in transactions:
        q = users[user]

        while q and time - q[0] > 60:
            q.popleft()

        q.append(time)

        if len(q) > 3 :
            flagged.append(user)
    
    return flagged    
        
```

## Rate Limiter


### Design

Allow: 100 requests/min

```python
#queue = deque()

class RateLimiter:
    def __init__(self):
        self.queue = deque()
    
    def rate_limiter(self,timestamp: int) -> bool:
        while self.queue and timestamp - self.queue[0] > 60:
            self.queue.popleft()
        
        if len(self.queue)  > 100:
            return False

        self.queue.append(timestamp)

        return True
```
